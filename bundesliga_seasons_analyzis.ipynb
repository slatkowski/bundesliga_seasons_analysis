# Data loading

To unpack our dataset I used zipfile function and after this was used classic function to read data - pd.read_csv. 

```python
with zipfile.ZipFile('bulidata.csv.zip', 'r') as file:
    file.extractall('')
    
df = pd.read_csv('bulidata.csv')
df
```

Next was used info() function to see information about entries and Dtype of columns. Also I looking for any null values by .isnull().any() and dataset have 2 columns with null values. This is in viewer and away team columns. 

```python
df.info()
print('____________________________')
print(df.isnull().any())
```

Change the capital letters in columns name to lower letters is done only for comfortable of data manipulation.

```python
df.columns = df.columns.str.lower()
df.columns
```
